```python
import covasim as cv

pfizer = cv.vaccinate_prob(vaccine='pfizer', days=40, prob=0.5)
az = cv.vaccinate_prob(vaccine='az', days=40, prob=0.5)

sim = cv.Sim(
    pars={
        'pop_size':5000000,
        'pop_infected': 10000,
        'pop_type': 'hybrid',
        'location': 'Australia',
        'start_day': '2021-09-21',
        'end_day': '2021-12-31',
        'rand_seed': 17
    },
    interventions=[pfizer, az],
    variants=cv.variant('p1', days=1),
    use_waning=True
)
sim.run()
sim.plot()
```
